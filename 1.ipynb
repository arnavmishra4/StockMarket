```python
import pandas as pd 
from binance import BinanceSocketManager
from sqlalchemy import create_engine
from binance.client import Client
import numpy as np
```

```python
client = Client()
```

```python
bsm = BinanceSocketManager(client)
```

```python
socket = bsm.trade_socket('BTCUSDT')
# enter the context manager
await socket.__aenter__()
# receive a message
msg = await socket.recv()
print(msg)
# exit the context manager
await socket.__aexit__(None, None, None)
```

```python
def datatransformation(msg):
    try:
        df = pd.DataFrame({'Time': msg['E'], 'Price': msg['p']}, index=[0])
        df.Price = df.Price.astype(float)
        df.Time = pd.to_datetime(df.Time, unit='ms')
        return df
    except Exception as e:
        # Handle the case where there is an issue with the data
        print(f"Returning DataFrame with NaN values due to exception: {e}")
        return pd.DataFrame({'Time': [pd.NaT], 'Price': [np.nan]})
```

```python
engine = create_engine('sqlite:///D:\PersonalProjects/BitcoinStream.db')
```

```python
current_event = pd.Series(pd.to_datetime(0))
while True:
    await socket.__aenter__()
    msg = await socket.recv()
    df = datatransformation(msg)
    if df.Time.values > current_event.values:
        current_event = df.Time
        df.to_sql('BitcoinStream',engine,if_exists= 'append',index=False)
```
